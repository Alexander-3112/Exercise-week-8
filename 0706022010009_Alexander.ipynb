```python
#load dataset
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Alexander-3112/Exercise-week-8/main/Property%20Dataset%20-%20Clustering%20-%20Data%20Rumah.csv")
df
```

```python
#missing value
from random import choices
import numpy as np
arah =["Utara","Selatan","Barat","Timur"]
fill = choices(arah, k = 83)
for i in df['Arah Hadap Rumah'].isnull():
    df["Arah Hadap Rumah"].replace([np.nan], arah[1], inplace = True)
df['Jumlah Kamar Mandi'].replace([np.nan],df['Jumlah Kamar Mandi'].mode()[0],inplace = True)  
posisi =["Tusuk Sate","Tikungan/Hook","Tusuk Belakang/Membelakangi Jalan","Kantong Belakang/Ngantong","Standard","Cul De Sac/Ujung Gang Buntu"]
fill2 = choices(posisi, k = 116)
for i in df['Posisi Rumah'].isnull():
    df["Posisi Rumah"].replace([np.nan], posisi[1], inplace = True)
df['Area Category'].replace([np.nan],df['Area Category'].mode()[0],inplace = True)
df['Nama Daerah'].replace([np.nan],df['Nama Daerah'].mode()[0],inplace = True)
df['Luas Tanah (m2)'].replace([np.nan],df['Luas Tanah (m2)'].mode()[0],inplace = True)
df['Luas Bangunan (m2)'].replace([np.nan],df['Luas Bangunan (m2)'].mode()[0],inplace = True)
df['Jumlah Kamar'].replace([np.nan],df['Jumlah Kamar'].mode()[0],inplace = True)
df['Tingkat/Lantai '].replace([np.nan],df['Tingkat/Lantai '].mode()[0],inplace = True)
df['Harga Penawaran (dari Owner)'].replace([np.nan],df['Harga Penawaran (dari Owner)'].mode()[0],inplace = True)
df['Terjual/Belum'].replace([np.nan],df['Terjual/Belum'].mode()[0],inplace = True)
lebar =["< 1 Mobil","1-2 Mobil","> 2 Mobil"]
fill3 = choices(lebar, k = 42)
for i in df['Lebar Jalan Depan Rumah (ROW)'].isnull():
    df["Lebar Jalan Depan Rumah (ROW)"].replace([np.nan], lebar[1], inplace = True)
```

```python
#labelencoder
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
df["Area Category"] = lb.fit_transform(df["Area Category"])
df["Nama Daerah"] = lb.fit_transform(df["Nama Daerah"])
df["Harga Penawaran (dari Owner)"] = lb.fit_transform(df["Harga Penawaran (dari Owner)"])
df["Terjual/Belum"] = lb.fit_transform(df["Terjual/Belum"])
df["Arah Hadap Rumah"] = lb.fit_transform(df["Arah Hadap Rumah"])
df["Posisi Rumah"] = lb.fit_transform(df["Posisi Rumah"])
df["Lebar Jalan Depan Rumah (ROW)"] = lb.fit_transform(df["Lebar Jalan Depan Rumah (ROW)"])
```

```python
# scale dataset with standard scaler
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
df_scale = scale.fit_transform(df)
df_scale
```

```python
#kmeans
from sklearn.cluster import KMeans
kmeans = KMeans(init="random",n_clusters=5,max_iter=300,random_state=42)
model = kmeans.fit_predict(df_scale)
model
df['K-MEANS'] = model
df
```

```python
# set the parameter for evaluating K
import matplotlib.pyplot as plt
wcss=[]
X2 = df_scale
# check the inertia for each K
for k in range(1,11):
    kmeans = KMeans(n_clusters= k, init="random")
    kmeans.fit(X2)
    wcss.append(kmeans.inertia_)  
plt.plot(range(1,11),wcss, linewidth=2, color="blue")
plt.xlabel("K Value")
plt.xticks(np.arange(1,11,1))
plt.ylabel("WCSS")
plt.show()    
```
